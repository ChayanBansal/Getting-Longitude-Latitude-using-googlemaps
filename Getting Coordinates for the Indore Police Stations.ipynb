```python
import pandas as pd

df = pd.read_csv("IndorePS.csv")
df
```

```python
df=df+", Indore, Madhya Pradesh, India"
df
```

```python
import googlemaps
gmaps_key = googlemaps.Client(key="#")
```

```python
df["LAT"] = None
df["LON"] = None
```

```python
for i in range(0,len(df),1):
    geocode_result = gmaps_key.geocode(df.iat[i,0])
    try:
        lat = geocode_result[0]["geometry"]["location"]["lat"]
        lon = geocode_result[0]["geometry"]["location"]["lng"]
        df.iat[i, df.columns.get_loc("LAT")] = lat
        df.iat[i, df.columns.get_loc("LON")] = lon
    except:
        lat = None
        lon = None
```
